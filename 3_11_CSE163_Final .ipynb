# **Cleaning Data**

## Cleaning Industries Dataset

This cell removes all unneccessary columns in the Industries dataset

```python
# Columns to keep
col_keep = ["GeoName", "Description", "Unit", "1997", "1998", "1999", "2000", "2001", "2002"]

industries_data = industries_data[col_keep]
industries_data
```

This cell removes rows that are regions of the United States, rather than state names or the entire US. This cell also removes rows with NaN values

```python
# Rows to remove from dataframe
remove_geoname = ["New England", "Mideast", "Great Lakes",
                  "Plains", "Southeast", "Southwest", "Rocky Mountain",
                  "Far West"]

for name in remove_geoname:
  industries_data= industries_data.drop(industries_data[industries_data["GeoName"] == name].index)

industries_data = industries_data.dropna()
industries_data
```

```python
# Making a year column and gdp column so that it is easier to merge datasets when asnwering questions
years_col = ["1997", "1998", "1999", "2000", "2001", "2002"]
year_gdp_list = []

for year in years_col:
  for gdp_value in industries_data[year]:
    year_gdp_list.append((year, gdp_value))

year_gdp_df = pd.DataFrame(year_gdp_list, columns=["year", "gdp"])

name_list = []
desc_list = []

for num_of_years in range(len(years_col)):
  for name in industries_data["GeoName"]:
    name_list.append(name)
  for desc in industries_data["Description"]:
    desc_list.append(desc)

name_df = pd.DataFrame(name_list, columns=["name"])

desc_df = pd.DataFrame(desc_list, columns=["description"])

industries_data = name_df.join(desc_df)

industries_data = industries_data.join(year_gdp_df)

industries_data
```

Is GDP increasing or decreasing on a national and state level? What does the trend look like when compared to polarization levels?

What is the trend?

```python
q1_data = pd.concat([cleaned_gdp, industries_data, clean_polarization])
q1_data
```

```python
# Groups data by country
grouped_country_df = q1_data.groupby('country')

increasing_country = 0
decreasing_country = 0

# Iterates over each country and finds trend
for country, group_df in grouped_country_df:
    if group_df['gdpppp'].iloc[-1] > group_df['gdpppp'].iloc[0]:
        increasing_country += 1
    else:
        decreasing_country += 1

print("total increasing is", increasing_country)
print("total decreasing is", decreasing_country)
```

```python
q1_state = q1_data[q1_data["description"] == "All industry total"]
q1_state["gdp"] = q1_state["gdp"].astype(float)
q1_state = q1_state.dropna(subset=["name"])
q1_state
```

```python
# Groups data by state
grouped_state_df = q1_state.groupby('name')

increasing_state = 0
decreasing_state = 0

# Iterates over each state (including US + District of Columbia) and finds trend
for state, group_df in grouped_state_df:
    if group_df['gdp'].iloc[-1] > group_df['gdp'].iloc[0]:
        increasing_state += 1
    else:
        decreasing_state += 1

print("total increasing is", increasing_state)
print("total decreasing is", decreasing_state)
```

## Motivation-
Our motivation for pursuing this topic is to learn more about the relationship between various industries in the United States and political polarization. Also, we want to learn about how politics can possibly affect other areas of society, specifically the economy.


## Data Setting-
We used three datasets from two different data sources: Kaggle and the World Bank Data Catalog.
From the World Bank Data Catalog, we used this dataset: https://datacatalog.worldbank.org/search/dataset/0047367/Globalization-and-Income-Distribution-Dataset-1975-2002, which describes the GDP per country for all countries in the world, from years 1975-2002. From this dataset, we used the contcod (country codes), GDP, and the year columns. 

Next, from Kaggle, we used this dataset: https://www.kaggle.com/datasets/justin2028/political-polarization-in-us-congress?resource=download which describes the political polarization levels in the US Congress from the years, 1879-2021. We used the chamber, year, and party.mean.diff.d1 columns. The chamber column indicated the proportion of Democrats/Republicans in either the House or the Senate and the party.mean.diff.d1, which indicates political polarization, column took the average number of Democrats and Republicans in the House and Senate and calculated the difference between the averages. When the difference is closer to 1, the House and Senate are more polarized on views and when the difference is closer to 0, the House and Senate have a more equal number of Democrats and Republicans.

Lastly, from Kaggle again, we used this dataset: https://www.kaggle.com/datasets/davidbroberts/us-gdp-by-state-19972020 which describes GDP per state from the years 1997-2020. From this dataset, we used the name, description, year, and gdp columns. The name column indicates different US states as well as the US as a whole, and the description column indicates the different industries that contribute to the United States gdp. 
Our datasets all have varying year ranges, but the years we have in common are 1997-2002. This is advantageous because we can hone in on a particular time frame and dive deeper into the questions. However, we will be unable to find trends over the years within the data. 

## Method-
The methodology for this project is to examine three datasets obtained from the World Bank, Voteview, and the Bureau Of Economic Analysis. The datasets will be joined/merged together by year. The years of study will be between 1997 and 2002 because these are the only years of overlap between all datasets.
